```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from operator import itemgetter
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder, LabelEncoder
```

Загрузить таблицу и вычилсить промежуточные значения

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data', header=None)

sig = lambda x: 1.0/(1.0+np.exp(-x))
cnt = df[0].value_counts()
p0 = cnt['p']/cnt.sum()
z0 = np.log(p0) - np.log(1.0 - p0)
```

Посчитать число ядовитых и съедобных грибов для некоторых признаков и визуализировать это

```python
fig, sub = plt.subplots(2, 3, figsize=(10,7))

for i in range(1, 6):
    df.groupby(i)[0].value_counts().unstack().fillna(0).plot(kind='bar', ax=sub.flatten()[i])
```

Вычислить вероятность съедобности по каждому признаку и сделать предсказание на основе этого 

```python
pred_df = pd.DataFrame({})
prob_df = pd.DataFrame({})

counted_df = (df.groupby(i)[0].value_counts().unstack().fillna(0) for i in range(1, df.columns.size))
prob_df = pd.concat(counted_df, keys=range(1, df.columns.size))
prob_df['prob'] = (prob_df['p']+1)/(prob_df['p']+prob_df['e']+2)
prob_df['pred'] = prob_df['prob'].map(lambda x: 'e' if x<0.5 else 'p')

prob_df
```

Построить решающую функцию по каждой переменной на основе частот

reset_index(drop=True) меняет индексацию в таблице на обычные номера. Это нужно для нахождения ошибок предсказания, но кидает ошибку, если функция вызывается не на таблице

```python
mush_prob_predict = lambda idx, val: prob_df['pred'][idx][val]#.reset_index(drop=True)

# mush_prob_predict(1, 'x')
mush_prob_predict(1, df[1])
```

Вычислить точность каждого решения (функции по каждому параметру)

```python
mush_prob_errors = lambda idx: len(mush_prob_predict(idx, df[idx]).reset_index(drop=True).compare(df[0]))
mush_prob_precision = lambda idx: 1 - mush_prob_errors(idx) / len(df)

# mush_prob_precision(1)
mush_precisions = list((i, mush_prob_precision(i)) for i in range(1, df.columns.size))
print(max(mush_precisions, key=itemgetter(1))) # itemgetter(1) используется, чтобы найти максимум из кортежа по 2-му элементу
mush_precisions
```

Используем многочленный наивный байесовский классификатор из sklearn
Необходимо заметить, что MultinomialNB работает только на float, так что преобразуем значения-буквы в числа с помощью их ASCII кода

```python
nm_model = CategoricalNB()

# enc = OrdinalEncoder()
# nm_model.fit(c_to_int(df.iloc[:, 1:]), df.iloc[:, 0])
# nm_predictions = nm_model.predict(c_to_int(df.iloc[:, 1:]))
enc = LabelEncoder()
c_to_int = lambda data: enc.fit_transform(data)
nm_model.fit(df.iloc[:, 1:].apply(c_to_int), df.iloc[:, 0])

nm_predictions = nm_model.predict(df.iloc[:, 1:].apply(c_to_int))

# True if prediction was right, False otherwise
correctness = nm_predictions == df.iloc[:, 0]
nm_precision = len(correctness[correctness])/len(correctness) # correctness[correctness] - фильтрует так, чтобы остались только True
nm_precision # 0.9560561299852289
```

```python
my_naive_df = prob_df.drop(['pred', 'prob'], axis=1)
my_naive_df['val_prob_p'] = (my_naive_df['p'] + 1) / (cnt['p'] + 2)
my_naive_df['val_prob_e'] = (my_naive_df['e'] + 1) / (cnt['e'] + 2)
my_naive_df.loc[2]
```
