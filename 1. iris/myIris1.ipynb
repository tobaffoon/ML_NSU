Imports

```python
import numpy as np
import pandas as pd
from sklearn import datasets
```

Загрузка данных

```python
iris = datasets.load_iris()
```

Создание дата фрейма

```python
iris_frame = pd.DataFrame(iris.data,columns = iris.feature_names)
iris_frame
```

Добавление имён классов 

```python
iris_frame['name'] = np.array(iris['target_names'][x] for x in iris['target'])
iris_frame.iloc[::10,::1]
```

Корреляция всех свойств попарно по классам

```python
iris_frame.groupby('name').corr()
```

Корреляция по всей таблице

```python
iris_frame.iloc[:, :-2].corr() # -2 значит кроме последнего столбца (имён) 
```

Визуализация распределение по классам для всех пар свойств

```python
import seaborn as sns

#sns.pairplot(iris_frame.iloc[:,:5], hue = 'target')
sns.pairplot(iris_frame, hue = 'name')
```
