### Посчитать значимость параметров

На значимость по чистоте параметров (feature_importances_) влияет то, сколько раз по ним происходит решения в деревьях, а также насколько чище (pure) становится решение после прохождения через узел с этим признаком по критерию Джини. Берётся среднее по всем деревьям.

Не самый точный показатель, так как он отдаёт предпочтения признакам, у которых множество возможных значений больше.

Также надо отметить, что в любом виде значимость параметров отражает их влияние на конкретную модель, то есть не всегда точна для не тренировочных данных

```python
named_impurity_importances = list(zip(X.columns, clf.feature_importances_)) # скрепляем значимость параметров с их именами
sorted(named_impurity_importances,key=itemgetter(1), reverse=True) # сортируем по их значимости
```

Permutation feature importance - уменьшение в точности модели при перемешивании значений только признака.

```python
# n_repeats - сколько раз значения параметров перемешано
# ATTENTION: долго считается
r = permutation_importance(clf, X, Y, n_repeats=30, random_state=0)
```

```python
# скрепляем мат ожидание и среднеквадратическое отклонение значимости параметров с их именами
named_impurity_importances = list(zip(X.columns, r.importances_mean, r.importances_std)) 
sorted(named_impurity_importances,key=itemgetter(1), reverse=True) # сортируем по их значимости
```

## Задание 3. Построить зависимость качества решения от числа вершин дерева

### Посчитать точности для глубин [1,6]

Число вершин = 2^max_depth 

```python
# ATTENTION: долго считается
points = []
for depth in range(1,7):
    clf_r = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=depth, random_state=0).fit(X,Y)
    p = clf_r.predict(X)
    q = (p==Y).sum()/Y.shape[0]
    points.append([2**depth, q])
```

### График

```python
_, ax = plt.subplots(1, 1, figsize=(12,6))
points = np.array(points)
ax.plot(points[:, 0], points[:, 1])
# ax.plot([1 for i in range(1, 65)], linestyle='dashed')
```

### Функции для применения скользящего контроля

```python
def crossval(X,Y,model,n_splits):
    kf = KFold(n_splits)
    scor = [] # список точностей для перебираемых разбиений
    for train, test in kf.split(X):
        X_train, X_test, Y_train, Y_test = X.iloc[train], X.iloc[test], Y[train], Y[test]
        clf = model.fit(X_train,Y_train)
        p = clf.predict(X_test)
        q = (p==Y_test).sum()/Y_test.shape[0] # точность для рассматриваемого разбиения 
        scor.append(q)
    return np.array(scor)

def crossval_staged(X,Y,model,n_splits):
    kf = KFold(n_splits)
    scor = []
    for train, test in kf.split(X):
        X_train, X_test, Y_train, Y_test = X.iloc[train], X.iloc[test], Y[train], Y[test]
        clf = model.fit(X_train,Y_train)
        q = np.array([(p==Y_test).sum()/Y_test.shape[0] for p in clf.staged_predict(X_test)])
        scor.append(q)
    return np.array(scor)
```

### Теперь для скользящего контроля

```python
# ATTENTION: очень долго считается
points = [] # список пар количеств вершин с точностями, полученными при скользящем контроле
for depth in range(1,7):
    model_r = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=depth, random_state=0)
    q = crossval(X, Y, model_r, 5)
    points.append((2**depth, q)) 
```

```python
points
```

### График

В конце оно даже начало уменьшаться

```python
_, ax = plt.subplots(1, 1, figsize=(12,6))

# Отображаем полученные 5 точностей (5 разделений в скользящем контроле было) в среднее и превращаем в массив
vertices_accuracy = np.array(list(map(lambda tupl: (tupl[0], tupl[1].mean()), points)))
print(vertices_accuracy)
ax.plot(vertices_accuracy[:, 0], vertices_accuracy[:, 1]) # 0 - число вершин; 1 - точность
# ax.plot([1 for i in range(1, 65)], linestyle='dashed')
```

## Задание 4. Построить зависимость качества решения от числа деревьев

### Зависимость bias от числа шагов

```python
q = np.array([(p==Y).sum()/Y.shape[0] for p in clf.staged_predict(X)])
plt.figure(figsize=(12,6))
plt.plot(q)
plt.plot([1 for i in range(1001)], linestyle='dashed')
```

### Предположительно оценка variance с помощью скользящей выборки

```python
# ATTENTION: очень долго считается
q_cv = crossval_staged(X,Y,clf,5)
```

```python
print(q_cv.shape) # массив, в котором 5 (по параметру KFold) массивов по 1000 точностей (для количеств деревьев от 1 до 1000)
q_cv
```

```python
# Plot crossvalidated
plt.figure(figsize=(12,6))
plt.plot(q_cv.mean(axis=0))

# Plot regular
q = np.array([(p==Y).sum()/Y.shape[0] for p in clf.staged_predict(X)]) # здесь глубина 3
plt.plot(q)

plt.plot([1 for i in range(1001)], linestyle='dashed')
```

### Совместим с перебором по глубине

```python
## ATTENTION: исполняется дольше всего (вроде дольше 10 минут)
depths_accs = []
for depth in range(1,7):
    model_r = GradientBoostingClassifier(n_estimators=1000, learning_rate=0.1, max_depth=depth, random_state=0)
    q = crossval_staged(X, Y, model_r, 5) # получить результат для определенной глубины размером (5, 1000) - 5 разделений, от 1 до 1000 деревьев
    r = q.mean(axis=0) # mean преобразует результат в массив размером 1000 - среднее по 5 разбиениям
    depths_accs.append((2**depth, r))
    
depths_accs
```

### Графики

```python
plt.figure(figsize=(12,6))
for (vertices, accs) in depths_accs:
    plt.plot(accs, label=f'{vertices} vertices')

plt.legend(loc="lower right")
# plt.plot([1 for i in range(1001)], linestyle='dashed')
plt.show()
```

```python
best_params = []
for (vertices, accs) in depths_accs:
    best_n_trees = np.argmax(accs)
    best_params.append((vertices, best_n_trees, accs[best_n_trees])) # массив кортежей (число вершин, количество деревьев, дающее наибольшую точность, получаемая точность)

print(best_params)
print(max(best_params, key=itemgetter(2)))
```

### Поиск по глубине и по числу деревьев

```python
depths_accs = []
for depth in range(1,9):
    n_trees_acc = []
    for n_trees in range(1, 31):
        model_r = RandomForestClassifier(n_estimators=n_trees, max_depth=depth, random_state=0)
        q_cv = crossval(X, Y, model_r, 5)
        n_trees_acc.append((n_trees, q_cv.mean(axis=0)))
    depths_accs.append((2**depth, n_trees_acc))
    
depths_accs
```

```python
plt.figure(figsize=(12,6))
for (vertices, accs) in depths_accs:
    plt.plot(*zip(*accs), label=f'{vertices} vertices')

plt.legend(loc="lower right")
plt.show()
```

```python
best_params = []
for (vertices, accs) in depths_accs:
    best_n_trees = max(accs, key=itemgetter(1))
    best_params.append((vertices, best_n_trees[0], best_n_trees[1])) # массив кортежей (число вершин, количество деревьев, дающее наибольшую точность, получаемая точность)

print(best_params)
print(max(best_params, key=itemgetter(2)))
```
